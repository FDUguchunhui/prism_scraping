```python
def extract_patient(file, start):
    N_ROWS = 20
    tbl = pd.read_excel(file, skiprows=start, nrows=N_ROWS)
    return tbl
```

Define a helper function that will remove empty columns and the last column with name "Scaled score" which is a duplicate of the first column, and unify the column name format

```python
def clean_colnames(dataframe):
    # get the column index that will be remained
    # remove a column if all rows is null 
#     col_idx = dataframe.columns.drop(list(dataframe.filter(regex=re.compile("unname", re.IGNORECASE))))
    df = dataframe.dropna(axis=1, how='all', inplace=False)
    df = df.iloc[:, :-1]
    # remove whitespace in both ends
    df.columns = df.columns.str.strip()
    # replace whitespace of any length with a single "space" chacacter
    df.columns = list(map(lambda colname: re.sub('\s+', ' ', colname), df.columns))
    # convert all columns to integer type
    df = df.fillna(pd.NA).astype('Int64')
    return df
```

Define a helper function that will do a inner join a list of dataframe iteratively on "Scaled score"

```python
def merge_on_score(dataframes):
    
    df = functools.reduce(lambda left,right: pd.merge(left,right,on='Scaled score',
                                            how='outer'), dataframes)
    df.columns = ['Scaled score', 'WAIS-R FSIQ', 'WAIS FSIQ', 'AIR', 'CAT ERROR','TRAIL B', 'THUR FLU', 'FAS', 'ANIMALS', 'DIGIT SPAN', 'DIGIT SPAN 2', 'PASAT', 'DIGIT ERROR', 'SSHOR RHYM', 'SPCH PERC', 
                  'DIGIT SYMB', 'DIGIT SYMB 2', 'TRAIL A', 'DIGIT TIME', 'BLOCK DESGN', 'BLOCK DESGN 2', 'TPT TIME', 'SPAT REL', 'WAIS-R VOCAB', 'WAIS VOCAB', 'APHAS SCRN', 'BSTN NAME', 'BDAE COMP', 'PIAT RECOG', 'PIAT COMP', 'PIAT SPELL', 
                  'STORY-T1', 'STORY LEARN', 'STORY RECALL', 'STORY LOSS', 'CVLT-1', 'CVLT-5', 'CVLT-TOT', 'CVLT-SDFR', 'CVLT-LDFR', 'TPT MEM', 'TPT LOC', 'FIGUR-T1', 'FIGUR LEARN', 'FIGUR RECALL', 'FIGUR LOSS', 
                  'TAP DH', 'TAP NDH', 'TAP BH', 'GRIP DH', 'GRIP NDH', 'PEG DH', 'PEG NDH', 'TPT DH', 'TPT BH', 'SPR', 'SPL', 'SP TOT', 'TFR R', 'TFR L']
    return df
```

# Extract the table meta-data from title

```python
meta_data = []
for start_index in table_start_idx:
    title_str = workbook.iloc[start_index].tolist()[0]
    gender = re.search('sex:([\s\S]+?)education', title_str, re.IGNORECASE).group(1).strip()
    
    # some level has education as a single point not range
    education = re.search('education:([\s\S]+?)years', title_str, re.IGNORECASE).group(1).strip()

    edu_from_match = re.search('^([\s\S]+?)-', education, re.IGNORECASE)
    if edu_from_match is not None:
        edu_from = int(re.search('^([\s\S]+?)-', education, re.IGNORECASE).group(1).strip())
        edu_to = int(re.search('-([\s\S]+?)$', education, re.IGNORECASE).group(1).strip())
    else:
        edu_from = int(education)
        edu_to = int(education)
    
    
    age = re.search('age:([\s\S]+?)years', title_str, re.IGNORECASE).group(1).strip()
    age_from = float(re.search('^([\s\S]+?)-', age, re.IGNORECASE).group(1).strip() + '.0')
    age_to = float(re.search('-([\s\S]+?)$', age, re.IGNORECASE).group(1).strip() + '.11')
    
    race = re.search('race:(.+)[\S]+', title_str, re.IGNORECASE).group(1).strip()
    
    meta_data.append({
        'age_from': age_from,
        'age_to': age_to,
        'edu_from': edu_from,
        'edu_to': edu_to,
        'race': race,
        'gender': gender
    })
```

```python
def get_JSON(test_info):
    '''
    test_info: a dict that contains three key (string) -values (string) pairs that provides name of neuro_test_name, neuro_test_field_name, column_name
    '''
    target_data = []
    for meta_table, norms_tbl in zip(meta_data, norms_tbls):
        result = norms_tbl.to_json(double_precision=0, orient='table',index=False)
        parsed_data = json.loads(result)['data']

        # parse the scale score and corresponding field score into list of dicts, each element is a row in target ".rake" file
        scale_scores = [{'scaled_score': value for key, value in tbl.items() if key == 'Scaled score'} for tbl in parsed_data]
        raw_scores = [{'raw_score': value for key, value in tbl.items() if key == test_info['column_name']} for tbl in parsed_data]
        table_data = [{**scale_score, **raw_score} for scale_score, raw_score in zip(scale_scores, raw_scores)]

        # combine the two parts of information
        for row_data in table_data:
            target_data.append({**meta_table, **row_data})

            
    # add neuro_test_name and neuro_test_field for all rows that belongs to that test 
    # this field adding is for all rows for the test (test-level)
    for target_row in target_data:
            target_row.update({'neuro_test_name': test_info['neuro_test_name'], 'neuro_test_field_name': test_info['neuro_test_field_name']})
            
    key_list = ['age_from', 'age_to', 'edu_from', 'edu_to', 'race', 'gender', 'neuro_test_name', 'neuro_test_field_name', 'scaled_score', 'raw_score']     
        
        # sort fields
#         ordered_target_data = []
#         for target_row in target_data:
#              ordered_target_data.append(OrderedDict((k, target_row[k]) for k in key_list))
        
    return target_data
```
