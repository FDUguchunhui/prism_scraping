```python
import pandas as pd
import json
import re
```

```python
stroop_table_V = pd.read_excel('data/Stroop_table.xlsx', sheet_name='table5', dtype=int)
```

```python
stroop_table_V
```

```python
table5_long = pd.melt(stroop_table_V, id_vars=['WORD'], value_vars=range(2, 114, 4)).sort_values(['WORD', 'variable'])
```

```python
table5_long.columns = ['total_correct_word', 'total_correct_color', 'predicted_interference_score']
```

```python
result = table5_long.to_json(orient='table',index=False)
```

```python
data_json = json.loads(result)['data']
```

```python
json_string = json.dumps(data_json, indent=4)
```

```python
json_string
```

```python
rake_string = re.sub('"', '', json_string)
# rake_string = re.sub('(?<=[^}][,{])\n', ' ', rake_string)
```

```python
with open('output/stroop_table_V.rake', 'w') as f:
    f.write(rake_string)
```
